```python
from google_play_scraper import Sort, reviews
import pandas as pd
import matplotlib.pyplot as plt
```

```python
app_id = "com.traveloka.android"

result, _ = reviews(
    app_id,
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=10000
)

print(f"Jumlah ulasan yang diambil: {len(result)}")

df = pd.DataFrame(result)

display(df.head())
```

```python
df = df[['userName', 'content', 'score', 'at', 'replyContent']]

df['date'] = df['at'].dt.date
df['time'] = df['at'].dt.time

display(df.head())
```

```python
df.to_csv("traveloka_reviews.csv", index=False, encoding='utf-8-sig')
print("Data berhasil disimpan")
```
